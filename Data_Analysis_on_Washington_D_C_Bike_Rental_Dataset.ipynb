#**Checking** **if**  **more** **bikes** **are** **rented** **by** **registered** **users** **on** **working** **or** **non**-**working** **days** ? **Does** **the** **answer** **differ** **for** **non**-**registered** **users** ? **Is** **the** **answer** **the** **same** **for** **both** **years**  

```python
df.workingday.value_counts()
```

```python
for year in [2011, 2012]:
    print(f"Analysis for the year {year}:")

    yearly_data = df[df['year'] == year]

    regworkday = yearly_data.groupby('workingday')['registered'].sum()
    print(f"Registered Users (Working vs Non-working days):\n{regworkday}")

    casworkday = yearly_data.groupby('workingday')['casual'].sum()
    print(f"Casual Users (Working vs Non-working days):\n{casworkday}")

    if regworkday[1] > regworkday[0]:
        regmore = "working days"
    else:
        regmore = "non-working days"

    if casworkday[1] > casworkday[0]:
        casualmore = "working days"
    else:
        casualmore = "non-working days"

    print(f"More registered users rent bikes on {regmore}.")
    print(f"More casual users rent bikes on {casualmore}.")
    print("\n")
```

**Registered Users**: More rentals on working days **(458,670)** than non-working days **(167,492)**.

**Casual Users**: More rentals on non-working days **(83,099)** than working days **(72,718)**.

**Conclusion**: Registered users prefer working days, while casual users prefer non-working days.

#**Checking which months in the year 2011 and 2012 have the highest and the lowest total number of bikes rented** ?

```python
for year in [2011, 2012]:
    print(f"Analysis for the year {year}:")

    yearly_data = df[df['year'] == year]

    monthly_rentals = yearly_data.groupby('month')['count'].sum()

    highest_month = monthly_rentals.idxmax()
    highest_rentals = monthly_rentals.max()
    lowest_month = monthly_rentals.idxmin()
    lowest_rentals = monthly_rentals.min()

    print(f"Highest rentals: Month {highest_month} with {highest_rentals} rentals.")
    print(f"Lowest rentals: Month {lowest_month} with {lowest_rentals} rentals.\n")
```

#**Conclusion**

**Analysis for the year 2011**:

The highest number of rentals occurred in **July**, with **92,848** rentals.
The lowest number of rentals occurred in **January**, with **23,552** rentals.

**Analysis for the year 2012**:

The highest number of rentals occurred in **September**, with **133,425** rentals.
The lowest number of rentals occurred in **January**, with **56,332** rentals.

#**Finding which type of weather have the highest and lowest mean of the hourly total rentals count?**

**Weather type**

1: 'Clear, Few clouds, Partly cloudy, Partly cloudy'.

2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist'.

3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds'.

4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'

```python
# Group by 'weather' and 'hour' to calculate mean count of rentals for each weather-hour combination
weather_hour = df.groupby(['weather', 'hour'])['count'].agg('mean').reset_index()

# Now group by 'weather' to calculate the mean rental count across all hours for each weather type
weather_rentals = weather_hour.groupby('weather')['count'].agg('mean')

# Finding the weather type with the highest and lowest mean rentals
highest_weather = weather_rentals.idxmax()
highest_mean = weather_rentals.max()

lowest_weather = weather_rentals.idxmin()
lowest_mean = weather_rentals.min()

# Display the results
print(f"The weather type with highest mean of hourly total rentals is '{highest_weather}' with an average of {highest_mean:.2f} rentals per hour.")
print(f"The weather type with lowest mean of hourly total rentals is '{lowest_weather}' with an average of {lowest_mean:.2f} rentals per hour.")
```

#**Conclusion**

The weather type with the highest mean of hourly total rentals is **'Clear, Few clouds, Partly cloudy'** with an average of **203.30** rentals per hour.

The weather type with the lowest mean of hourly total rentals is **'Light Snow, Light Rain with Thunderstorm and Scattered Clouds'** with an average of **111.45** rentals per hour.


#**Calculating the correlation between the hourly total rentals count and all thenumerical columns in the dataset and checking which column has the highest correlation with the total rentals count**

```python
corrdf = df.select_dtypes(include=['number'])

correlation = corrdf.corr()

# Extracting and sorting the correlation values for 'count'
countc = correlation['count']
sortcor = countc.sort_values(ascending=False)

print("Correlation of 'count' with numerical columns:")
print(sortcor)
```

```python
highestcorr = sortcor.index[1]  #Index is 1 Because the index 0 is the count column
highestvalue = sortcor.iloc[1]

print(f"The column with the highest correlation with 'count' is '{highestcorr}' with a correlation value of {highestvalue}.")
```

#**Conclusion**
The column with the highest correlation with 'count' is **'registered'**, with
a correlation value of **0.9709**, indicating a very strong positive relationship between the number of registered users and the total bike rentals

#**Creating a new categorical column called day_period, which can take fourpossible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning,12-6: afternoon, 6-24:evening**

```python
bins = [0, 6, 12, 18, 24]
labels = ['night', 'morning', 'afternoon', 'evening']
df['day_period'] = pd.cut(df['hour'], bins=bins, labels=labels, right=False)
```

```python
df.head(10)
```

#**Generating a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table**

```python
pivot_table = df.pivot_table(
    values='count',
    index='day_period',
    columns='workingday',
    aggfunc='mean',
)
print(pivot_table)
```

#**Conclusion**

On non-working days **(workingday = 0)**, the highest mean rental count is in the afternoon **(371.02 rentals)**, followed by evening **(180.18 rentals)** and morning **(157.77 rentals)**. The night period has the lowest mean rental count **(44.05 rentals)**.

On working days **(workingday = 1)**, the highest rentals occur in the afternoon **(266.20 rentals)**, followed by evening **(251.05 rentals)** and morning **(234.35 rentals)**. The night period has the lowest mean rental count **(16.22 rentals)**
